### 6. List the movies released between 1967 and 1995.

```python
cursor = coll.find(
    {'$and': [
      {'year': {'$gte': 1967}}, 
      {'year': {'$lte': 1995}}
    ]},
    {'_id', 'title', 'year'}
)

for doc in cursor:
  print(doc)
```

### 7. List the the movies released between 1967 and 1995, by displaying only title, year, director’s last name sorted by year.

```python
cursor = coll.find(
    {'$and': [
        {'year': {'$gte': 1967}},
        {'year': {'$lte': 1995}}
    ]},
    {'title', 'year', 'director.last_name'}
).sort('year', 1)

for doc in cursor:
  print(doc)
```

### 8. Find the number of movies by country. (Extra credit - Hint: use aggregation)

```python
cursor = coll.aggregate([
    {'$group': { '_id': '$country', 'count': { '$sum': 1 } }}
])

for doc in cursor:
  print(doc)
```

### 9.  Find the number of movies by country and actor. Limit the output to actors that have appeared in more than 2 movies.  (Extra credit - Hint: use aggregation and unwind in the aggregation pipe)

```python
movies_by_country = coll.aggregate([
    {'$group': { '_id': '$country', 'count': { '$sum': 1 } }}
])
movies_by_actor = coll.aggregate([
    {'$unwind': '$actors'}, 
    {'$group': 
     {'_id': '$actors._id', 'count': { '$sum': 1 }}
    },
    {'$match': {'count': {'$gt': 2 }}}
])

print('Movies by country')
for doc in movies_by_country:
  print(doc)
print()
print('Movies by actor')
for doc in movies_by_actor:
  print(doc)
```

### Join movies collection and box_office collection

```python
cursor = movies.aggregate([
  {'$lookup': {'from': 'box_office', 'localField': '_id', 'foreignField': "movie_id", "as": "box_office"}}
])

for doc in cursor:
  print(doc)
```

### Get the average rating by genre. Sort by average rating.

```python
cursor = movies.aggregate([ 
  {'$lookup': {'from': 'ratings', 'localField': '_id', 'foreignField': "movie_id", "as": "ratings"}},
  {'$unwind': "$ratings"},
  {'$group': {'_id': '$genre', 'avg_rating': {'$avg': {'$toDouble': '$ratings.rating'}}}},
  {'$sort': {'avg_rating': 1}}
])

for doc in cursor:
  print(doc)
```

### Get the average box office gross by genre. Sort by average sales.

```python
cursor = movies.aggregate([ 
  {'$lookup': {'from': 'box_office', 'localField': '_id', 'foreignField': "movie_id", "as": "box_office"}}
  {'$unwind': "$box_office"},
  {'$group': {'_id': '$genre', 'avg_box_office': {'$avg': {'$toDouble': '$box_office.gross'}}}},
  {'$sort': {'avg_box_office': 1}}
])

for doc in cursor:
  print(doc)
```
