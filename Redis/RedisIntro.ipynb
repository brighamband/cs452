```python
!pip install redis-server redis

import redis
import redis_server
!$redis_server.REDIS_SERVER_PATH --daemonize yes

age_r = redis.Redis(host='localhost', port=6379, db=1)
languages_r = redis.Redis(host='localhost', port=6379, db=2)

age_r.ping()
languages_r.ping()


# set up AGE key-value store
age_r.set('John', 12)
age_r.set('Paul', 10)
age_r.set('George', 8)
age_r.set('Ringo', 16)
age_r.set('Russell M Nelson', 98)
age_r.set('Boss Baby', 2)
age_r.set('Nebuchadnezzar', 2542)


# setup language key-value store
languages_r.sadd('John', 'English', 'French', 'Spanish')
languages_r.sadd('George', 'English', 'French')
languages_r.sadd('Ringo', 'English')
languages_r.sadd('Paul', 'English', 'French', 'German')

languages_r.sadd('Russell M Nelson', 'English','Mandarin', 'Spanish', 'French', 'Portuguese', 'Russian')
languages_r.sadd('Boss Baby', 'English', 'Baby')
languages_r.sadd('Nebuchadnezzar',"Babylonian")
```

# Q1) Age value for President Nelson

```python
print(int(age_r.get("Russell M Nelson")))
```

# Q2) New key value pair for Abraham Lincoln, Age 213

```python
age_r.set('Abraham Lincoln', 213)
print(age_r.get("Abraham Lincoln"))
```

#Q3) Decrement Abraham Lincoln's age by 157

```python
age_r.decr('Abraham Lincoln', 157)
print(age_r.get("Abraham Lincoln"))
```

```python
keys = age_r.keys()
max_age = 0
oldest_man = keys[0]
for key in keys:
  age = int(age_r.get(key))
  if age > max_age:
    max_age = age
    oldest_man = key

print(oldest_man, max_age)
  
```
