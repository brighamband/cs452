```
from pyspark.sql.functions import col
```

```
# 1. Write code to define the schema and then read in the dataset

df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("dbfs:/FileStore/shared_uploads/brighamband@gmail.com/us_counties.csv")
df.show(10)
```

```
# 2. Write code to find the county with the most deaths
df2 = df.orderBy(col("deaths").desc())
df2.select("county", "deaths").show(1)
```

```
# 3. Write code to find the county with the most cases
df3 = df.orderBy(col("cases").desc())
df3.select("county", "cases").show(1)
```

```
# 4. Write code to find the total number of deaths in Utah county
df4 = df.filter(df["county"] == "Utah")
df4.groupBy("county").max("deaths").show()
```

```
# 5. Write code to find the death rate for each state and sort the states by death rate
df5 = df.groupBy("state").max("deaths", "cases")
df5 = df5.withColumn("death rate", df5["max(deaths)"] / df5["max(cases)"])
df5.orderBy(col("death rate").desc()).select("state", "death rate").show()
```
